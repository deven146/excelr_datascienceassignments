**3. Standardization of Numerical Variables**

```python
# Standardization of numerical variables
# Numerical columns to standardize
num_cols = ['Volume', 'AveragePrice', 'Total Sales Value', 'Discount Rate (%)',
             'Discount Amount', 'Net Sales Value']
```

```python
# Save original for comparison
original_data = data[num_cols].copy()
original_data
```

```python
# Standardize numerical columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_standardized = data.copy()
df_standardized[num_cols] = scaler.fit_transform(data[num_cols])
```

```python
# Visualize before and after standardization
fig, axes = plt.subplots(2, len(num_cols), figsize=(20, 8))
fig.suptitle("Before and After Standardization", fontsize=16)

for i, col in enumerate(num_cols):
    sns.histplot(original_data[col], ax=axes[0, i], kde=True, color='skyblue')
    axes[0, i].set_title(f'Original: {col}')
    sns.histplot(df_standardized[col], ax=axes[1, i], kde=True, color='lightgreen')
    axes[1, i].set_title(f'Standardized: {col}')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Conversion of Categorical Data into Dummy Variables (One-Hot Encoding)
# Categorical columns to encode
categorical_cols = ['Day', 'City', 'BU', 'Brand', 'Model']
categorical_cols
```

```python
# Apply one-hot encoding
df_encoded = pd.get_dummies(df_standardized, columns=categorical_cols)
```
